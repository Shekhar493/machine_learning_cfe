```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
categories = [ "Rent", "Food", "Transport", "Entertainment"]
amounts = [ 800, 300, 150, 100 ]
```

```python
plt.bar(categories, amounts)
plt.show()
```

```python
plt.bar(categories, amounts)

plt.title("Monthly Budget Breakdown")
plt.xlabel("Categories")
plt.ylabel("Amounts")

plt.grid(True)

plt.show()
```

```python
plt.bar(categories, amounts, color='Red')

plt.title("Monthly Budget Breakdown")
plt.xlabel("Categories")
plt.ylabel("Amounts")

# plt.grid(True)

plt.show()
```
