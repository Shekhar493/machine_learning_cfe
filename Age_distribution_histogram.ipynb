```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ages = [18, 19, 19, 20, 20, 20, 21, 21, 22, 22, 22, 22, 23, 23, 24, 25]

plt.hist(ages, bins=5, color='lightcoral', edgecolor='black')
plt.title('Student Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()
```
