```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline

print("Seaborn version:", sns.__version__)
```

```python
tips = sns.load_dataset('tips')

print(tips.head())
print("\nDataset shape:", tips.shape)
print("\nColumn info:")
print(tips.info())
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Bill Amount vs Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',palette='Set1', hue='sex')
plt.title('Bill vs Tip (colored by sex)')
plt.show()
```
